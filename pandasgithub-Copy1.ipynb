### Simple Series

### Built-in Functions 

```python
data= [10,20,30,40,50]                     
a= pd.Series(data, index = ['a','b','c','d','e'])
print(a)
print(a.sum())    # total
print(a.mean())   # average
print(a.max())    # maximum
print(a.min())    # minimum
```

## upload csv file

```python
a=pd.read_csv("netflix_titles.csv")
a.head()
```

**CSV file ko read karo:-** CSV file ko pandas DataFrame mein read karne ke liye **pd.read_csv()** use karte hain:

**'filename.csv'** ki jagah aap apni file ka exact path ya file name dena.

```python
a.info()
```

```python
a.columns = a.columns.str.strip()
```

```python
a.isnull().sum()
```

```python
a['date_added']=pd.to_datetime(a['date_added'],errors="coerce")
```

```python
duration_split = a['duration'].str.extract(r'(\d+)\s*(\w+)')            ## Regular expressions (regex) are a powerful tool for extracting specific patterns from text data.
a['duration_num'] = pd.to_numeric(duration_split[0],errors ='coerce')
a['duration_unit']=duration_split[1]  
```

```python
a.isnull().sum()
```

```python
a['duration_unit'].value_counts()
```

```python
a['duration_unit'] = a['duration_unit'].replace({'Seasons': 'Season'})
```

```python
a['duration_unit'].value_counts()
```

```python
a['duration_num'] = a.groupby('duration_unit')['duration_num'].transform(lambda x: x.fillna(x.median()))   #missingnvalue (nan) ko fill karna
```

```python
# df.groupby('duration_unit') --> sab rows ko unke duration_unit ke hisaab se group karo. season group, min group
#  ['duration_num'] -- means group ke andar sirf number column.
# The transform() function in pandas is a powerful tool that allows you to apply a function to each element of a DataFrame or Series,
# returning a result with the same shape as the original data. 
# This is particularly useful when you need to apply a transformation function 
# that returns a result for each element while maintaining the structure (rows and indexes) of the original DataFrame.
###############################################################################################

# import pandas as pd

# t = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})
# t['A_transformed'] = t['A'].transform(lambda x: x * 2)
# t['B_transformed'] = t['B'].transform(lambda x: x * 2)
# print(t)
```

```python
a.loc[a['duration'].isna() & (a['type'] == 'Movies'),'duration_unit '] ='min'
# Jitni rows me duration missing (NaN) hai, Aur type = Movie hai Un rows ke duration_unit me 'min' fill kar do
a.loc[a['duration'].isna() & (a['type'] == 'TV Show'),'duration_unit '] ='season'

# loc[] = label-based selection -> Rows ko condition se select karta hai, Aur un rows me kisi column ki value change ya read kar sakte ho.
# df.loc[df['age'] > 18, 'status'] = 'Adult'
# isna() = checks missing values (NaN, empty) -> give result in True or False
```

```python
a.head()
```

```python
a = a.drop(columns=['duration'])
```

```python
a.tail()
```

```python
# # # # # # Option A  Fill missing with "Unknown"
# # # # # df['rating'] = df['rating'].fillna('Unknown')
# # # # Option B  Fill missing with mode (most common rating)
# # # most_common = df['rating'].mode()[0]
# # # df['rating'] = df['rating'].fillna(most_common)
# # Option C  Create a missing flag + fill (BEST for ML)
# df['rating_missing'] = df['rating'].isna().astype(int)
# df['rating'] = df['rating'].fillna('Unknown')
```

```python
a['rating_missing'] = a['rating'].isna().astype(int)  # True-1 , False-0
a['rating'] = a['rating'].fillna('Unknown')

# df['rating_missing'] = df['rating'].isna().astype(int) , # True-1 , False-0
# df['rating'] = df['rating'].fillna('Unknown')
```

```python
a.isnull().sum()
```

```python
a.duplicated().sum()
```

```python
median_date = a['date_added'].median()
a['date_added'] =a['date_added'].fillna(median_date)
```

```python
a.isnull().sum()
```

```python
a['country_missing'] = a['country'].isna().astype(int)
a['country'] = a['country'].fillna("Unknown")
```

```python
a.isnull().sum()
```

```python
a['director_missing'] = a['director'].isna().astype(int)  # True - 1 , False - 0
a['cast_missing'] = a['cast'].isna().astype(int)
```

```python
a['director'] = a['director'].fillna("Unknown")
a['cast'] = a['cast'].fillna("Unknown")
a['duration_unit'] = a['duration_unit'].fillna("Unknown")
```

```python
a['duration_num'] = a['duration_num'].fillna("Unknown")
```

```python
a.isnull().sum()
```
