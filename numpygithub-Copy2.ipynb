## Numpy Array

### 1D Array
*Ye 1-Dimensional array hai Sirf ek row hoti hai.*
#### **print(arr)** = *Array ke sab elements ko print karta hai.*

```python
arr = np.array([11,12,13,14,15,16,17])
print(arr)
```

### 2D Array (Matrix)
 *Ye 2D array hai (rows + columns)*

```python
arr1 = np.array([[12,13,14,11],[15,124,15,15],[21,4,54,54]])
print(arr1)
```

### 3D Array (Single Block)
 *Ye 3-D array hai Matlab: 1 block, 4 rows, 3 columns*

```python
arr2 = np.array([[[12,12,12],[11,42,52],[43,45,98],[3,5,5]]])
print(arr2)
```

### 3D Array (Multiple Blocks)
*Jab ek array ke andar multiple 2D matrices (layers) hoti hain, to use Multiple Blocks bolte hain.*

```python
arr3 = np.array([[[45,22,45,15],[58,54,55,45]],
                 [[45,84,43,54],[75,45,65,58]]])
print(arr3)
```

**size:-** *total elements count karta hai ,Formula: blocks * rows * columns = 2 * 2 * 4 = 16*

### Mathematical Functions  :-



```python
arr = np.array([10, 20, 30, 40])
print(arr.sum())
```

**sum :-** *array ke andar jitni bhi values hain, unka sum ek hi baar me mil jaata hai.*

```python
arr = np.array([10, 20, 30, 40])
print(arr.mean())
```

**mean :-** *NumPy ka function hai ,jo array ka average (Arithmetic Mean) nikalta hai. Mean = (sab values ka sum) / (total values)*

```python
arr = np.array([10, 45, 23, 90, 12])
print(arr.max())
```

**max :-** *jo array ke andar sabse badi (maximum) value nikalta hai.*

```python
arr = np.array([10, 45, 23, 90, 12])
print(arr.min())
```

**min :-** *jo array ke andar sabse chhoti (minimum) value nikalta hai.*

```python
arr = np.array([1, 4, 9, 16, 25])
print(np.sqrt(arr))
```

**sqrt :-** *jo array ke har element ka square root nikalta hai.*

### Matrix Functions :- 

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = np.dot(a, b)
print(result)
```

**dot:-** *Dot product (sum of products), Matrix multiplication*

```python
arr= np.array([[1, 2, 3],[4, 5, 6]])
print(arr.T)
```

**transpose(T):-** *jo matrix ka rows aur columns swap karta hai.*

```python
arr = np.array([10, 45, 23, 99, 34])
print(np.argmax(arr))
```

**argmax():-** *NumPy ka ek function hai jo array ke andar sabse badi (maximum) value ka index position batata hai.*

```python
arr = np.array([10, 45, 23, 5, 34])
print(np.argmin(arr))
```

**argmin():-** *NumPy ka ek function hai jo array ke andar sabse chhoti (minimum) value ka index position batata hai.*

```python
arr = np.array([10, 20, 30, 40, 50])
print(np.std(arr))
```

**std():-** *NumPy ka ek function hai jo Standard Deviation calculate karta hai.Data mean (average) se kitna spread / faila hua hai, uska measure hota hai.*

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr.flatten())
```

**flatten():-** *NumPy ka function hai jo multi-dimensional array (2D, 3D, etc.) ko 1D array me convert kar deta hai.*

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr.ravel())
```

**ravel():-** *Array ko flat banata hai, lekin zyada tar cases me original array ka “view” deta hai.*

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr.reshape(2, 3))
```

**reshape():-** *NumPy ka function hai jo array ka shape change karta hai.Data same rehta hai, sirf rows–columns ka structure change hota hai.*

```python
arr = np.array([1, 2, 3, 4])
print(np.diag(arr))
```

**diag():-** NumPy ka function hai jo diagonal elements ke saath kaam karta hai.

## Difference Between NumPy Array and Python List

 *Python me data store karne ke liye List aur NumPy Array dono use hote hain.Dono ka kaam same lag sakta hai, lekin performance, usage aur power me kaafi difference hota hai.*

*Numpy perform faster for larger-scale mathematical operations while list slower for numerical computations due to per-element overheead.*

*Numpy supports advanced mathematical functions,linear algebra, broadcasting and multi-dimensional arrays list offers general-purpose methods like sort,count,and reverse.*

```python
list1 = [1, 2, 3, 4]
array1 = np.array([1,2,3,4])
print("List Addition: ",list1 + list1)
print("Numpy array Addition : ", array1 +array1)
```

*List repeat the values and give output .Numpy perfrom addition and give the sum of values in output.*
