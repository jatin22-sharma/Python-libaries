## Barplot

A barplot shows the average (mean) value of a column grouped by categories.

```python
sns.barplot(data=df, x='day', y='tip', hue='sex')
plt.title("Day VS Tip")
plt.show()
```

**sns.barplot :-** create the barplot graph. "df" is variable. On x-axis = "day" and on y-axis = "tip" and hue is used to compare to categories inside one plot.

**plt.title :-** Used for add title to the graph.

**plt.show :-** used for display the graph.

## Countplot

A countplot is used to show the count of observations in each category of a categorical variable.

```python
sns.countplot(data=df, x='day') 
plt.show()
```

**sns.countplot :-** create the countplot graph. On x-axis = "day" (count how many people visited each day).

## Histplot

A histplot is used to show the distribution of numerical data (how values are spread).

```python
sns.histplot(data=df, x='total_bill', hue='sex')
plt.show()
```

**sns.histplot :-** create histplot graph. In data taking variable "df" and in x-axis = "total_bill" and "hue" is used to compare to categories inside one plot. 

## KDE = Kernel Density Estimate

It's basically a smooth curve that shows the probability distribution of data.

Its used to understand:

  Shape of the data
 
  Peak values
 
  Skewness
 
  Spread
 
  Multi-modal distribution

 KDE is like a smoother, elegant version of a histogram.

```python
sns.kdeplot(data=df, x='total_bill', fill=True)
plt.show()
```

**sns.kdeplot :-** create kdeplot graph. In data taking variable "df" and in x-axis = "total_bill" and "fill" is used to fill color.

## boxplot

It helps to understand distribution and outliers in numerical data.

A boxplot shows:

Median

Quartiles (Q1, Q3)

Spread of data

Outliers

```python
sns.boxplot(data=df, x='day', y='total_bill', hue='sex')
plt.show()
```

**sns.boxplot :-**  create boxplot graph. In data taking variable "df" and in x-axis = "day" and in y-axis ="total_bill" and "hue" is used to compare to categories inside one plot.

## Violinplot

 A violin plot shows:

 Full distribution of the data (like KDE)
 
 Median
 
 Quartiles
 
 Density shape (wide = many values, thin = fewer values)

 It is basically:
 
 Boxplot + KDE plot merged together
 
 Extremely useful for understanding:
 
  Skewness
 
  Peaks
 
  Compare distributions between categories
 
  Range + Density simultaneously

```python
sns.violinplot(data=df, x='day', y='total_bill', hue='sex')
plt.show()
```

**sns.violinplot :-** create violinplot graph. In data taking variable "df" and in x-axis = "day" and in y-axis ="total_bill" and "hue" is used to compare to categories inside one plot.

```python
# When is Violin Plot Better Than Boxplot?

# | Boxplot                        | Violinplot                  |

# | ------------------------------ | --------------------------- |

# | Shows summary statistics only  | Shows full distribution     |

# | Cannot see where data is dense | Density is clearly visible  |

# | Simple, clean                  | More detailed               |

# | Good for outliers              | Good for shape + comparison |
```

## Scatterplot

 A scatter plot helps you understand:

 Relationship between two numeric variables
 
 Positive / Negative correlation
 
 Clusters
 
 Trends
 
 Patterns
 
 Outliers
 
 This is the MOST commonly used plot in ML EDA.

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', size='size', style='smoker')
plt.show()
```

**sns.scatterplot :-** create scatterplot graph. In data taking variable "df" and in x-axis = "total_bill" and in y-axis ="tip" and "hue" is used to compare to categories inside one plot.
"size" Controls the size of the points.
"style" Controls the shape/marker of the points.

## Lineplot

 A line plot helps you understand:

 Trend over time
 
 Changes between values
 
 Increasing / decreasing patterns
 
 Seasonality
 
 Line plots are the default choice for:
 
  Date columns
 
  Time series
 
  Daily/Monthly values
 
  Any variable that changes in order

```python
sns.lineplot(data=df, x='total_bill', y='tip', marker='o')
plt.show()
```

**sns.lineplot :-** create lineplot graph. In data taking variable "df" and in x-axis = "total_bill" and in y-axis ="tip" and "marker" the shape of the point drawn at each data point on the line.

## Heatmap

 Heatmaps are used to visualize correlation between numeric columns.
 
 Correlation helps you understand:
 
 Which feature affects the target
 
 Multicollinearity (two features too similar)
 
 Which columns are useless
 
 Which features are important for ML
 
 Correlation values range from:
 
 +1  Strong positive
 
 -1  Strong negative

 0  No relationship

```python
corr = df.corr(numeric_only=True)
corr
```

Create Correlation Matrix

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

**sns.heatmap :-** create heatmap graph.  annot=True  shows numbers and cmap="coolwarm"  color theme.

## Pairplot

 A pairplot creates:

 Histogram for every variable

 Scatter plot between every pair of numeric variables

 Optionally different colors (hue) for categories

 It's like pressing one button and getting full relationship analysis of your dataset.

**Importance of pairplot**

  Shows correlation visually
 
  Shows cluster patterns
 
  Shows linear/non-linear relationships
 
  Shows outliers
 
  Shows distribution
 
  Helps identify feature importance for ML
 
 This is used heavily in ML EDA (Exploratory Data Analysis).

```python
sns.pairplot(df)
plt.show()
```

**sns.pairplot :-** create pairplot graphs. This automatically selects numeric columns.
