## Line plot

Create a simple line plot. 

```python
x = np.arange(1,10)
y = np.arange(11,20)
plt.plot(x, y, label='Sine Wave')
plt.xlabel('Number upto 1-9')
plt.ylabel('Number upto 11-19')
plt.title('Simple Line Plot')
plt.grid()
plt.legend()
plt.show()
```

* **x** variable me NumPy array banaya jata hai jisme values 1 se start hoti hain aur 9 par end hoti hain (10 include nahi hota).
* **y** variable me NumPy array banaya jata hai jisme values 11 se start hoti hain aur 19 tak jati hain.
* **plt.plot(x, y, label="Sine Wave")** x aur y values ka simple line plot banata hai aur line ko label deta hai.
* **plt.xlabel("Number upto 1-9")** x-axis ko label provide karta hai.
* **plt.ylabel("Number upto 11-19")** y-axis ko label provide karta hai.
* **plt.title("Simple Line Plot")** graph ka title set karta hai.
* **plt.grid()** graph ke background me grid show karta hai.
* **plt.legend()** label (Sine Wave) ko graph par display karta hai.
* **plt.show()** final line plot graph ko screen par show karta hai.


## scatter plot

Do variables ke beech relationship ko samajhne ke liye graph ke through unki values ko visually show karna

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='r' , marker='*')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

* **x** variable NumPy ka use karke 0 se 1 ke beech ke 20 random numbers generate karta hai.
* **y** variable bhi NumPy ka use karke 0 se 1 ke beech ke 20 random numbers generate karta hai.
* **plt.scatter(x, y, color="blue", marker="o")** x aur y values ka scatter plot banata hai jisme blue color aur o marker use hota hai.


## Bar chart

*  *Alphabets aur numbers ke beech relationship ko dikhane ke liye **bar graph** ka use kiya jata hai.
 Is graph me har alphabet ke saamne uska corresponding number **rectangular bars** ke form me represent hota hai.*



```python
categories = ['A', 'B', 'C', 'D' ,'E']
values = [3, 7, 1, 8 ,5.8]
plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

* **categories** variable me category names ki list store ki jati hai.
* **values** variable me numerical values ki list store ki jati hai.
* **plt.bar(categories, values, color="green")** ka use karke bar chart create hota hai jisme categories x-axis par aur values bars ki height hoti hain, aur bars ka color green hota hai.


## Histogram

Histogram numerical data ka distribution dikhata hai, jisme data ko ranges (bins) me divide karke unki frequency bars ke through represent ki jati hai.

```python
data = np.random.randn(1000)
plt.hist(data, bins=70, color='red', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

* **plt.hist(data, bins=70, color='red', edgecolor='black'):-** yeh histogram banata hai jisme 70 bins hote hain, bars red color ke aur black border ke saath hote hain.
* **plt.xlabel('Value'):-** x-axis ko Value label diya jata hai.
* **plt.ylabel('Frequency') :-** y-axis ko Frequency label diya jata hai.
  



# sub plot


```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(x, y, color='blue', linestyle='--')
ax[0].set_title('Sine Wave')
ax[1].bar(categories, values, color='red')
ax[1].set_title('Bar Chart')
plt.show()
```

* **fig, ax = plt.subplots(1, 2, figsize=(10, 4)):-** ek figure aur 2 side-by-side axes (subplots) create karta hai jiska size 10x4 inches hai.
* **ax[0].plot(x, y, color='blue', linestyle='--'):-** pehle subplot me x aur y values ka dashed blue line plot banata hai.
* **ax[0].set_title('Sine Wave'):-** pehle subplot ka title Sine Wave set karta hai.
* **ax[1].bar(categories, values, color='red'):-** dusre subplot me categories aur values ka red color bar chart banata hai.
* **ax[1].set_title('Bar Chart'):-** dusre subplot ka title Bar Chart set karta hai.
* **plt.show():-** dono subplots ko ek saath screen par display karta hai.


# Pie plot

```python
plt.subplot(2,2,4)  # 4th plot
labels = ["Python", "Java", "C++", "Other"]
sizes = [40, 30, 20, 10]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=['red','blue','green','yellow'])
plt.title('Pie Chart')
```


* Data: [40,30,20,10]  har slice ka size
* Labels: ["Python","Java","C++","Other"]  har slice ka naam
* autopct="%1.1f%%"`  percentage show kare 1 decimal ke saath
* colors=[...]  slice colors set kare
